# Projet : Détection d'anomalies sur Séries Temporelles

## 1. Détection de données anormales sur une série temporelle
Dans cette partie on simule des données de capteurs et on utilise le Z-Score pour identifier des comportements atypiques sur un signal.

$$Z = \frac{x - \mu}{\sigma}$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simulation de données
np.random.seed(13)
t = np.linspace(0, 100, 500)
signal = np.sin(t) + np.random.normal(0, 0.1, 500) #on ajoute du bruit
```

```python
# Ajout de points anormaux
signal[120] = 4.5
signal[300] = -3.8

df = pd.DataFrame({'Temps': t, 'Valeur': signal})
```

```python
# Détection par Z-Score glissant
window = 20 #comparaison avec la moyenne des 20 derniers points
df['Moyenne'] = df['Valeur'].rolling(window=window).mean()
df['STD'] = df['Valeur'].rolling(window=window).std()
df['Z_Score'] = (df['Valeur'] - df['Moyenne']) / df['STD']

# Seuil d'alerte (généralement 3 car 0,3% de chances que ce soit dû au hasard)
df['Anomalie'] = df['Z_Score'].abs() > 3
```

```python
# Visualisation
plt.figure(figsize=(12, 5))
plt.plot(df['Temps'], df['Valeur'], label='Signal Capteur', alpha=0.7)
plt.scatter(df[df['Anomalie']]['Temps'], df[df['Anomalie']]['Valeur'], 
            color='red', label='Anomalies détectées', zorder=5)
plt.title("Surveillance en temps réel : Détection d'écarts critiques")
plt.legend()
plt.show()
```

## 2. Détection de profils anormaux

On utilise ensuite le calcul de l'Erreur Quadratique Moyenne (MSE) pour identifier les courbes qui s'éloignent trop de la moyenne.

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_{réel} - Y_{moyen})^2$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulation de courbes de référence
np.random.seed(42)
x = np.linspace(0, 10, 100)
normal_curves = []

for i in range(50):
    # on ajoute du bruit
    noise = np.random.normal(0, 0.2, 100)
    curve = np.sin(x/2) + (x * 0.5) + noise 
    normal_curves.append(curve)
```

```python
# Ajout de courbes anormales
anomalies = [
    np.sin(x/2) + (x * 0.8),# Pente trop forte 
    np.sin(x/2) + (x * 0.2),# Pente trop faible 
    np.sin(x/2) + (x * 0.5) + 1.5 # Saut brusque
]

all_curves = np.array(normal_curves + anomalies).T
df = pd.DataFrame(all_curves)#chaque colonne est une courbe
```

```python
# Calcul du profil de référence (Moyenne et Écart-type)
df['Mean_Profile'] = df.iloc[:, :50].mean(axis=1)
df['Std_Profile'] = df.iloc[:, :50].std(axis=1)
```

```python
# Calcul du MSE entre chaque courbe et la moyenne
mse_scores = []
for col in df.columns[:-2]: # On exclut les colonnes Mean et Std
    mse = np.mean((df[col] - df['Mean_Profile'])**2)
    mse_scores.append(mse)

# Seuil de détection
threshold = np.mean(mse_scores) + 1.5 * np.std(mse_scores)
anomalous_indices = [i for i, score in enumerate(mse_scores) if score > threshold]
```

```python
# Visualisation
plt.figure(figsize=(12, 6))
# Tracer les courbes normales en gris
plt.plot(df.iloc[:, :50], color='gray', alpha=0.2, label='Normal' if i==0 else "")
# Tracer le profil moyen en noir
plt.plot(df['Mean_Profile'], color='black', linewidth=2, label='Profil Moyen (Référence)')
# Tracer les anomalies détectées en rouge
for idx in anomalous_indices:
    plt.plot(df[idx], color='red', linewidth=1.5, label=f'Anomalie {idx}')

plt.title("Détection de profils de serrage anormaux (Profile Monitoring)")
plt.xlabel("Angle / Temps")
plt.ylabel("Couple (N.m)")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()

print(f"Indices des courbes anormales détectées : {anomalous_indices}")
```
